### Breaking point.

```python
r_min_intersection = np.inf
phi_meet_intersection = 0

phi0 = 0 
phi1 = 0 

num_intersection = 0 
phi_meet_vec = []
r_meet_vec = []

for ii in range(len(p)):     
    for jj in range(ii, len(p) -1):
        if ii != jj:
            deltap= (p[ii] - p[jj])
            num = v_at_30_radii_equator[ii] * v_at_30_radii_equator[jj]
            den = v_at_30_radii_equator[ii] - v_at_30_radii_equator[jj]
            
            res =  (r[0].value- (deltap/omega_rot)*(num/den))* u.km

            if  r[0] <  res < r[-1]:        
                num_intersection += 1
                
                phi_meet = p[ii] - omega_rot*(res.value - r[0].value)/(v_at_30_radii_equator[ii])
                
                phi_meet_vec.append(phi_meet)
                r_meet_vec.append(res.to(u.AU).value)
                
                if res.value < r_min_intersection:        
                    r_min_intersection = res.value
                    phi_meet_intersection = phi_meet
                    phi0 = p[ii], ii
                    phi1 = p[jj], jj
```

# Shock Fitting - Whitham's Equal Area Rule
*Whitham's (1974) equal-area rule*: the multi-valued solution at radial distance from the Sun, $r$, the shock $s(r)$ is located by drawing a vertical line which cuts off equal area lobes. If we replace part of the multivalued solution interval with a shock, some mass will be removed and some mass will be added. Number conservation means that when we replace the multivalued solution by the discontinuous one, the area under the curve must remain the same.
Extensively, this Whitham equal-area rule argues that both the multivalued solution and the discontinuious solution satisfy the conservation law. Thus, for any breaking wave at $s(r)$, the area under each curve must be equal.

```python
def area_compute(index1, index2, x_shifted, u, s,  method="trap"):
    """ function to compute area between a vertical line x=s(t) and the multivalued function u(x, t)"""
    if method=="trap":
        return np.trapz(x_shifted[index1:index2+1] - s, x=u[index1:index2+1])
    if method=="simps":
        return scipy.integrate.simpson(x_shifted[index1:index2+1], x=u[index1:index2+1])
    else:
        return None



def get_index(diff):
    """ get indexes where x=s(t) and the multivalued function u(x, t) intercept"""
    asign = np.sign(diff)
    signchange = (np.diff(asign) != 0)*1
    return np.where(signchange==1)[0]
```

```python
def get_s_opt(t_idx, x_vec, t_vec, a_shock, b_shock, a, b, initial_condition, n_search=int(1e5)):
    """ function to get the optimal s (shock) based on Whitham (1974) geometrical arguments """
    # shifted coordinates
    x_shifted = x_vec - (omega_rot/initial_condition)*(t_vec[t_idx] - t_vec[0])

    # shock location initial guess
    s_vec = np.linspace(a_shock, b_shock, n_search)
    area_diff = np.zeros(n_search)

    for s_idx, s in enumerate(s_vec):
        # difference between shifted location and current shock s
        diff = x_shifted - s
        # figure out where the two curves intersect
        index = get_index(diff)

        if len(index) != 3: 
            area_diff[s_idx] = np.inf

        else:
            # compute the area between the two curves
            area1 = np.abs(area_compute(index1=index[0], index2=index[1], x_shifted=x_shifted, u=initial_condition, s=s))
            area2 = np.abs(area_compute(index1=index[1], index2=index[2], x_shifted=x_shifted, u=initial_condition, s=s))
            area_diff[s_idx] = np.abs(area1 - area2)  
            
    if t_idx % 10 == 0 and len(get_index(x_shifted- s_vec[np.argmin(area_diff)])) == 3:       
        fig, ax = plt.subplots(figsize=(10, 5))
        ax.scatter((180/np.pi)*(x_shifted%b), initial_condition, s=2)
        ax.scatter((180/np.pi)*(s_vec[np.argmin(area_diff)]%b), initial_condition[get_index(x_shifted- s_vec[np.argmin(area_diff)])[0]], c="r")
        ax.scatter((180/np.pi)*(s_vec[np.argmin(area_diff)]%b), initial_condition[get_index(x_shifted- s_vec[np.argmin(area_diff)])[1]], c="r")
        ax.scatter((180/np.pi)*(s_vec[np.argmin(area_diff)]%b), initial_condition[get_index(x_shifted- s_vec[np.argmin(area_diff)])[2]], c="r")
        ax.axvline((180/np.pi)*(s_vec[np.argmin(area_diff)]%b), c="k", linewidth=1, label="approximate = " + str(round(s_vec[np.argmin(area_diff)], 3)))
        ax.set_xlabel("Carrington Londitude [Rad]")
        ax.set_ylabel("$v_{r}(\phi, r)$ [km/s]")
        ax.set_title(r"r = " + str(round(((t_vec[t_idx]*u.km).to(u.AU)).value, 3))+" AU \n $|A_{1} - A_{2}|$ = " + str(np.min(area_diff)))

        plt.legend()
        plt.show()

    return s_vec[np.argmin(area_diff)] 
```

```python
s_curve1 = np.ones(len(r)) * np.inf
s_curve2 = np.ones(len(r)) * np.inf
for r_idx, rr in enumerate(r.value):
    if rr > r_min_intersection + 1e-2:
        # s(r) -- compute for each r slice
        s_curve1[r_idx] = get_s_opt(t_idx=r_idx, x_vec=p, t_vec=r.value, a_shock=2.8, b_shock=4, 
                                   a=0, b=2*np.pi, initial_condition=v_at_30_radii_equator)
        
        s_curve2[r_idx] = get_s_opt(t_idx=r_idx, x_vec=p, t_vec=r.value, a_shock=4, b_shock=5, 
                                   a=0, b=2*np.pi, initial_condition=v_at_30_radii_equator)
```

# Shock curves (fitting)

```python
for ii in range(len(s_curve1)):
    if s_curve1[ii] != np.inf:
        n_start1 = ii 
        break

n_start2 = n_start1 + np.argmax(s_curve2[n_start1:])
```

```python
def func_asymp(t, a, b):
    return a*t + b 
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(15, 6))

popt = scipy.optimize.curve_fit(func_asymp, r[n_start1:].to(u.AU).value, (180/np.pi)*s_curve1[n_start1:])[0]
ax[0, 0].plot(r[n_start1:].to(u.AU).value, (180/np.pi)*s_curve1[n_start1:], ls="-", label="shock curve")
ax[0, 0].plot(r[n_start1:].to(u.AU).value, func_asymp(r[n_start1:].to(u.AU).value, *popt), ls="--", label="linear approximate")
_ = ax[0, 0].set_ylabel("$s_{1}(r)$ [Deg.]")
_ = ax[0, 0].set_xlabel("Heliocentric Distance [AU]")
_ = ax[0, 0].set_title("$s_{1}(r) = $" + str(round(popt[0], 3))+ "r + " + str(round(popt[1], 3)))
_ = ax[0, 0].legend()


ax[1, 0].plot(r[n_start1:].to(u.AU).value, np.abs((180/np.pi)*s_curve1[n_start1:] - func_asymp(r[n_start1:].to(u.AU).value, *popt)))
_ = ax[1, 0].set_ylabel("Abs. Error [Deg.]")
_ = ax[1, 0].set_xlabel("Heliocentric Distance [AU]")
_ = ax[1, 0].set_title("|$s_{1}(r)$ - linear fit|")


popt = scipy.optimize.curve_fit(func_asymp, r[n_start2:].to(u.AU).value, (180/np.pi)*s_curve2[n_start2:])[0]
ax[0, 1].plot(r[n_start2:].to(u.AU).value, (180/np.pi)*s_curve2[n_start2:], ls="-", label="shock curve")
ax[0, 1].plot(r[n_start2:].to(u.AU).value, func_asymp(r[n_start2:].to(u.AU).value, *popt), ls="--", label="linear approximate")
_ = ax[0, 1].set_ylabel("$s_{2}(r)$ [Deg.]")
_ = ax[0, 1].set_xlabel("Heliocentric Distance [AU]")
_ = ax[0, 1].set_title("$s_{2}(r) = $" + str(round(popt[0], 3))+ "r + " + str(round(popt[1], 3)))
_ = ax[0, 1].legend()

ax[1, 1].plot(r[n_start2:].to(u.AU).value, np.abs((180/np.pi)*s_curve2[n_start2:] - func_asymp(r[n_start2:].to(u.AU).value, *popt)))
_ = ax[1, 1].set_ylabel("Abs. Error [Deg.]")
_ = ax[1, 1].set_xlabel("Heliocentric Distance [AU]")
_ = ax[1, 1].set_title("|$s_{2}(r)$ - linear fit|")

plt.tight_layout()
```

# Collection of characteristic curves

```python
# initialize set of characteristic curves
char_curves = np.zeros((len(p), len(r)))

# loop over each point
for ii in np.arange(len(p)):
    for jj in np.arange(len(r)):
        x_loc = p[ii] - (omega_rot/v_at_30_radii_equator[ii])*(r[jj].value - r[0].value)
        if np.min(np.abs(s_curve1[jj] - x_loc)) < 1e-2:
            char_curves[ii, jj:] = s_curve1[jj:]*np.nan
            break
        if np.min(np.abs(s_curve2[jj] - x_loc)) < 1e-2 and jj > n_start2:
            char_curves[ii, jj:] = s_curve2[jj:]*np.nan
            break
        else:
            char_curves[ii, jj] = x_loc 
```

# Cross-Correlation

### Background
Cross-correlation is a mathematical operation that is commonly used in signal processing and pattern recognition to measure similarity of two functions as they are displaced from one another. For two finite discrete signals $f, g \in \mathbb{C}^{n}$, the uni-variate discrete *circular cross-correlation* ($\star$) is defined as

\begin{equation}\label{cc-dis-circular-definition}
    (f \star g)[\tau] \equiv \sum_{j=0}^{n-1} \overline{f[j]} g[(j + \tau)_{\text{mod} n}]
\end{equation}

where $\overline{f}$ denotes the complex conjugate of $f$, the bracket $[j]$ denotes the $j^{th}$ element of the signal, and $\tau \in \mathbb{Z}$ is the discrete displacement. In general, discrete circular cross-correlation can be extended to the multi-variate case for snapshots with $k \in \mathbb{Z}$ variables:

\begin{equation}\label{cc-dis-definition-multi-variate}
(f \star \overset {k}{\cdots } \star g)[\overline{\tau}] \equiv \sum_{j_{1}=-\infty}^{\infty} ... \sum_{j_{k}=-\infty}^{\infty}
\overline{f[j_{1}, ..., j_{k}]} g[(j_{1}+ \tau_{1})_{\text{mod} n_{1}}, (j_{2}+ \tau_{2})_{\text{mod} n_{2}}, ..., (j_{k} + \tau_{k})_{\text{mod} n_{k}}]
\end{equation}

where $\overline{\tau}= [\tau_{1}, \tau_{2},... ,\tau_{k}] \in \mathbb{Z}^{k}$ is the multi-variate discrete displacement. The optimal shift between the two discrete signals, $\bar{\tau}^{*}$, is found by the maximum of the cross-correlation, such that

\begin{equation}\label{cc-max}
    \bar{\tau}^{*} = {\text{argmax}_{\bar{\tau} \in \mathbb{Z}^{k}}} {(f \star \overset {k}{\cdots } \star g)[\overline{\tau}]}
\end{equation}

Once the optimal cross-correlation is found for all training snapshots by applying circular discrete cross-correlation between each snapshot and the initial condition, the shift function $\mathbf{c}(t)$ is found via polynomial curve fitting between the time increments $\left[t_{0}, t_{1}, ..., t_{f} \right]$ and corresponding optimal shift $\left[\tau^{\star}_{0}, \tau^{\star}_{1}, ..., \tau^{\star}_{f} \right]$.  The wave speed at the testing regime is approximated by extrapolating $\mathbf{c}(t)$ beyond the training regime. 

### Motivation to use to find spatial shifting in convective-dominated problems
When the functions correlate, the value of $(f \ast g)$ is maximized. This is because when peaks and troughs align, they make a positive contribution to the finite sum.

### Circular convolution theorem and cross-correlation theorem
Given sequences $x$ and $y_{N}$, where $y_{N}$ is N-periodic, then the cross-correlation of $x$ and $y_{N}$ is given by 
$$
x \ast y_{N} \equiv \mathcal{F}^{-1}\{ \overline{\mathcal{F}\{x\}} \cdot \mathcal{F}{\{y\}\}}
$$

where $\mathcal{F}$ denotes the Fourier transform, and $\overline{\mathcal{F}\{x\}}$ indicates the complex conjugate of $\mathcal{F}\{x\}$. This is analogous to the convolution theorem.

# Circular discrete cross-correlation

```python
def normalize(x):
    return (x - np.mean(x))/np.std(x)
```

```python
def periodic_corr_fft(x, y):
    """Periodic correlation, implemented using the FFT.
    x and y must be real sequences with the same length.
    """
    return ifft(fft(x).conj() * fft(y)).real

def periodic_corr_np(x, y):
    """Periodic correlation, implemented using np.correlate.

    x and y must be real sequences with the same length.
    """
    return np.correlate(x, np.hstack((y[1:], y)), mode='valid')
```

```python
res2 = periodic_corr_np(normalize(v1), normalize(v2))
res2 /= np.max(res2)
```

# Method of Characteristics vs. Cross-Correlation

```python
def compute_phi_shift_forward(p, r, v, omega=(2 * np.pi) / (25.38 * 86400), method=None):
    # initialize phi shift matrix. 
    phi_shift_mat = np.zeros((len(r), len(p)))
    
    # phi at index 0 is original phi grid 
    phi_shift_mat[0, :] = np.zeros(len(p))
    
    # delta r.
    dr = np.mean(r[1:] - r[:-1])
    
    # compute the phi shift for each idx in r. 
    for ii in range(len(r)-1):
        if method == "ballistic":
            phi_shift = -(omega/v[:, 0])*dr
        else:
            phi_shift = -(omega/v[:, ii])*dr
        phi_shift_mat[ii+1, :] = phi_shift_mat[ii, :] + phi_shift
        
    return phi_shift_mat
```
